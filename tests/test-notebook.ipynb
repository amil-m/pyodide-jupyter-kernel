## Test 6: File Mounting and CSV Loading

The `mount()` function loads files from your VSCode workspace into Pyodide's virtual filesystem.

```python
# Mount the test-data.csv file from workspace
# The file will be available at /test-data.csv in Pyodide's filesystem
pyodide_path = mount('test-data.csv')
print(f'File mounted at: {pyodide_path}')
print('Check the side panel to see mounted files!')
```

```python
# Load the CSV file using pandas
df_csv = pd.read_csv('/test-data.csv')

# Show basic info about the data
print(f"Loaded {len(df_csv)} rows and {len(df_csv.columns)} columns")
print(f"\nColumns: {', '.join(df_csv.columns)}")
print(f"\nData types:\n{df_csv.dtypes}")
print(f"\nSummary statistics:")
print(df_csv.describe())

# Display the dataframe
df_csv
```

```python
# Data manipulation examples
print("People with salary > $80,000:")
high_earners = df_csv[df_csv['Salary'] > 80000]
display(high_earners)

print("\nAverage salary by age group:")
df_csv['Age_Group'] = pd.cut(df_csv['Age'], bins=[20, 30, 40], labels=['20-30', '30-40'])
salary_by_age = df_csv.groupby('Age_Group')['Salary'].mean()
print(salary_by_age)
```

```python
# Create a visualization from the CSV data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart of salaries by person
ax1.bar(df_csv['Name'], df_csv['Salary'], color='skyblue')
ax1.set_title('Salary by Person')
ax1.set_xlabel('Name')
ax1.set_ylabel('Salary ($)')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(axis='y', alpha=0.3)

# Scatter plot of Age vs Salary
ax2.scatter(df_csv['Age'], df_csv['Salary'], color='coral', s=100, alpha=0.6)
ax2.set_title('Age vs Salary')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary ($)')
ax2.grid(True, alpha=0.3)

# Add labels for each point
for idx, row in df_csv.iterrows():
    ax2.annotate(row['Name'], (row['Age'], row['Salary']), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()
```
